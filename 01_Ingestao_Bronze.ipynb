```python
# Definir catálogo e schema padrão do projeto
spark.sql("USE CATALOG workspace")
spark.sql("USE SCHEMA default")

display(spark.sql("SELECT current_catalog(), current_schema()"))
```

```python
options = {
    "header": "true",
    "inferSchema": "true",
    "nullValue": "na",  # transforma 'na' em NULL
    "sep": ","
}

options
```

```python
# Caminho base do Volume com os CSVs do SCANIA
base_path = "/Volumes/workspace/default/scania_raw_data"


base_path
```

```python
display(dbutils.fs.ls(base_path))
```

```python
# TRAIN
path_train_ops   = f"{base_path}/train_operational_readouts.csv"
path_train_tte   = f"{base_path}/train_tte.csv"
path_train_specs = f"{base_path}/train_specifications.csv"

# VALIDATION
path_val_ops     = f"{base_path}/validation_operational_readouts.csv"
path_val_labels  = f"{base_path}/validation_labels.csv"
path_val_specs   = f"{base_path}/validation_specifications.csv"

# TEST
path_test_ops    = f"{base_path}/test_operational_readouts.csv"
path_test_labels = f"{base_path}/test_labels.csv"
path_test_specs  = f"{base_path}/test_specifications.csv"

paths = [
    path_train_ops, path_train_tte, path_train_specs,
    path_val_ops, path_val_labels, path_val_specs,
    path_test_ops, path_test_labels, path_test_specs
]

paths
```

```python
# TRAIN
df_train_ops_raw   = spark.read.options(**options).csv(path_train_ops)
df_train_tte_raw   = spark.read.options(**options).csv(path_train_tte)
df_train_specs_raw = spark.read.options(**options).csv(path_train_specs)

# VALIDATION
df_val_ops_raw     = spark.read.options(**options).csv(path_val_ops)
df_val_labels_raw  = spark.read.options(**options).csv(path_val_labels)
df_val_specs_raw   = spark.read.options(**options).csv(path_val_specs)

# TEST
df_test_ops_raw    = spark.read.options(**options).csv(path_test_ops)
df_test_labels_raw = spark.read.options(**options).csv(path_test_labels)
df_test_specs_raw  = spark.read.options(**options).csv(path_test_specs)




```

```python
print("=== TRAIN operational_readouts ===")
df_train_ops_raw.printSchema()
display(df_train_ops_raw.limit(5))

print("=== TRAIN tte ===")
df_train_tte_raw.printSchema()
display(df_train_tte_raw.limit(5))

print("=== TRAIN specifications ===")
df_train_specs_raw.printSchema()
display(df_train_specs_raw.limit(5))
```

```python
# TRAIN
df_train_ops_raw.write.format("delta").mode("overwrite").saveAsTable("bronze_train_operational_readouts")
df_train_tte_raw.write.format("delta").mode("overwrite").saveAsTable("bronze_train_tte")
df_train_specs_raw.write.format("delta").mode("overwrite").saveAsTable("bronze_train_specifications")

# VALIDATION
df_val_ops_raw.write.format("delta").mode("overwrite").saveAsTable("bronze_validation_operational_readouts")
df_val_labels_raw.write.format("delta").mode("overwrite").saveAsTable("bronze_validation_labels")
df_val_specs_raw.write.format("delta").mode("overwrite").saveAsTable("bronze_validation_specifications")

# TEST
df_test_ops_raw.write.format("delta").mode("overwrite").saveAsTable("bronze_test_operational_readouts")
df_test_labels_raw.write.format("delta").mode("overwrite").saveAsTable("bronze_test_labels")
df_test_specs_raw.write.format("delta").mode("overwrite").saveAsTable("bronze_test_specifications")


```
